* **Calculated are made for a year**
* **distribution is known(Passion) and Lambda is 0.1(average number of errors)**
* *QUEST* **What are the probabilities of no errors, 3 errors, and 5 errors?**

```python
from scipy.stats import poisson
```

```python
lambda_ = 0.1
rv = poisson(mu=lambda_)

```

```python
print(rv.pmf(0))
```

```python
print(rv.pmf(3))
```

```python
print(rv.pmf(5))
```
