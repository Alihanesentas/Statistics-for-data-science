Advertisements are given in various media and the click-through and conversion rates of the ads are tried to be optimized. For this purpose, the probability of clicking on the ad according to various scenarios in a certain medium is probablity to be calculated.


** DETAÄ°LS ** 
  * Only given ads one various media 
  * Distrubation and probablity of click-through : (0.01)
  * **QUEST : probablity of clicked 1,5,10 when ads seen 100 times**

```python
from scipy.stats import binom
```

```python
p = 0.01
n= 100
rv = binom(n, p)
print(rv.pmf(1))
print(rv.pmf(5))
print(rv.pmf(10))

```
