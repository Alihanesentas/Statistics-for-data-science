# Problem
**Before an investment/meeting, it is desired to determine the probabilities of next month's sales at certain values.**


## Details
* Distribution is normal
* Mean sales per month : 80k
* Standard Derivtion : 5k
* What is probablities to next month sale value too much 90K?

```python
from scipy.stats import norm
```

```python

1-norm.cdf(90, 80, 5) #Cumulative densty function
# 1-0.9772498680518208 = 0.02275013194817921
# 1 is a total probability 
```

```python
1-norm.cdf(70, 80, 5) # bigger than 70
```

```python
norm.cdf(73,80,5) # smaller than 73
```

```python
# between 85 and 90
norm.cdf(90,80,5) - norm.cdf(85,80,5)
```
